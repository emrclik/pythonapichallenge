```python
from citipy import citipy
import openweathermapy.core as owm
from config import api_key
import matplotlib.pyplot as plt
import requests
import pandas as pd
import numpy as np
import urllib
```

```python
url = 'http://api.openweathermap.org/data/2.5/weather?'
units = 'metric'

query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
latitudes = []
longitudes = []

#latituderange = (-90,90)
#longituderange = (-180,179)

latitudes = np.random.uniform(-90,90,[500])
longitudes = np.random.uniform(-180,179,[500])

city= citipy.nearest_city(latitudes[0],longitudes[0])

response_t = requests.get(query_url+city.city_name).json()
response_t
```

```python
coordinates = []
random_cities = []
coordinates = zip(latitudes, longitudes)

for a in coordinates:
    city = citipy.nearest_city(a[0], a[1]).city_name
    
    if city not in random_cities:
        random_cities.append(city)
len(random_cities)
```

```python
cities_data = []

print("Data Retr")
print("-----------------------------")

record_count = 1
set_count = 1

for i, city in enumerate(random_cities):
        
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 0

    city_url = query_url + urllib.request.pathname2url(city)

    print("Processing Record %s of Set %s | %s" % (record_count, set_count, city))
    print(city_url)

    record_count += 1

    try:
        
        city_weather = requests.get(city_url).json()

        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_date = city_weather["dt"]

        cities_data.append({"City": city, 
                          "Lat": city_lat, 
                          "Lng": city_lng, 
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})
    except:
        print("Not found")
        pass

print("-----------------------------")
print("Complete      ")
print("-----------------------------")
```

```python
pd_cities = pd.DataFrame(cities_data)
pd_cities.head()
```

```python
latitudes = pd_cities['Lat']
max_temperatures = pd_cities['Max Temp']

plt.scatter(latitudes,max_temperatures,marker ='o',color='red')
plt.title('Latitudes vs Max Temperatures')
plt.xlabel('Latitudes')
plt.ylabel('Temperatures')
plt.grid(True)
plt.show()
plt.savefig('Latitudes_MaxTemp')
```

```python
humidities = pd_cities['Humidity']

plt.scatter(latitudes,humidities,marker ='o',color='blue')
plt.title('Latitudes vs Humidities')
plt.xlabel('Latitudes')
plt.ylabel('Humidities')
plt.grid(True)
plt.show()
plt.savefig('Latitudes_Humidities')
```

```python
cloudiness = pd_cities['Cloudiness']

plt.scatter(latitudes,cloudiness,marker ='o',color='blue')
plt.title('Latitudes vs Cloudiness')
plt.xlabel('Latitudes')
plt.ylabel('Cloudiness')
plt.grid(True)
plt.show()
plt.savefig('Latitudes_Cloudiness')
```

```python
windspeed = pd_cities['Wind Speed']

plt.scatter(latitudes,windspeed,marker ='o',color='blue')
plt.title('Latitudes vs Wind Speed')
plt.xlabel('Latitudes')
plt.ylabel('Wind Speed')
plt.grid(True)
plt.show()
plt.savefig('Latitudes_Wind Speed')
```
